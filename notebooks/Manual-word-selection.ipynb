```python
def fill_word_frequency(row, dictionary):
    # define new rows containg word frequency, one row for each word in the dictionary
    bow = row['bow']
    for index in bow.keys():
        row[dictionary[index]] = bow[index]
    return row
```

```python
def get_impact(x):
    if x > 0.25:
        return 'pos'
    elif x < -0.25:
        return 'neg'
    else:
        return 'neutral'
```

```python
# df = innovative
# filename = '../processed/wc_innovative.csv'

# df = good
# filename = '../processed/wc_good.csv'

df = fb_all
filename = '../processed/wc_all_fb.csv'
```

```python
# build Dictionary of words in Post_text
dictionary = Dictionary(df['tokens'].values)
df['bow'] = df['tokens'].map(lambda x: dict(dictionary.doc2bow(x)))
df = df[['success', 'tokens', 'bow']]

df = df.apply(lambda x: fill_word_frequency(x, dictionary), axis=1)

words = []
for word in df.columns:
    if (word in ['success', 'tokens', 'bow']):
        continue
    dat = df[~df[word].isnull()]
    metric = dat['success']
    word_dat = {
     'word': word,
     'count': metric[~metric.isnull().values].count(), # no. of posts the word appeared in
     'success_mean': metric[~metric.isnull().values].mean(), # mean success value when word appears in post
    }
    words.append(word_dat)

word_summary = pd.DataFrame(words)[['word', 'count', 'success_mean']]
word_summary = word_summary.sort_values('success_mean')
word_summary['rank'] = word_summary['success_mean'].rank()
word_summary['score'] = (2 * (word_summary['rank'] - word_summary['rank'].min()) / (word_summary['rank'].max() -word_summary['rank'].min())) - 1
word_summary['impact'] = word_summary['score'].map(get_impact)
word_summary['type'] = 'used'
word_summary = word_summary[['word', 'score', 'impact', 'type']]

word_summary.to_csv(filename, index=False)
```

```python
emb, word2rank_dict = load_embedding_model('../utils/pl_pl_commoncrawl_v3.bin')
```

```python
def get_average_vector(words, emb):
    """
        Average the word vectors of a list of words.
    """
    words_in_emb = [emb[word] for word in words if word in emb.vocab.keys()]
    return np.mean(words_in_emb, axis=0)
```

```python
def cosine_similarity(a,b):
    """
        Compute the cosine similarity of a and b.
    """
    return np.dot(a,b) / ( (np.dot(a,a) **.5) * (np.dot(b,b) ** .5) )
```

```python
dimensions = pd.read_csv('../utils/dimensions2.csv')
dimensions['dimension'] = dimensions['pl'].map(lambda x: get_average_vector(tokenizer.tokenize(x), emb))
```

```python
innovative_wv = dimensions.loc[dimensions['en_label'] == 'Innovative', 'dimension'].iloc[0]
good_wv = dimensions.loc[dimensions['en_label'] == 'Feel Good', 'dimension'].iloc[0]
```

```python
volkswagen = 'volkswagen, vw'
volkswagen_wv = get_average_vector(tokenizer.tokenize(volkswagen), emb)
```

```python
top_1000_innovative = emb.similar_by_vector(innovative_wv, topn=1000)
top_1000_good = emb.similar_by_vector(good_wv, topn=1000)
#     print(top_1000_y_dim[:5])

for i in range(len(top_1000_innovative)):
    word, innovative_score = top_1000_innovative[i]
    vw_score = cosine_similarity(volkswagen_wv, emb[word])
    top_1000_innovative[i] = (word, innovative_score, vw_score)

for i in range(len(top_1000_good)):
    word, good_score = top_1000_good[i]
    vw_score = cosine_similarity(volkswagen_wv, emb[word])
    top_1000_good[i] = (word, good_score, vw_score)

top_1000_innovative.sort(key=lambda x: x[1]+2*x[2], reverse=True)
top_1000_good.sort(key=lambda x: x[1]+2*x[2], reverse=True)

# recommend_innovative = [x[0] for x in top_1000_innovative[:30]]
# recommend_good = [x[0] for x in top_1000_good[:30]]
```
